```python
import pandas as pd
import numpy as np
```

```python
df_history = pd.read_csv('../Dataset/Dataset_Version_7.csv')
df_recent = pd.read_csv('../Dataset/E0.csv')
df_ea_teams = pd.read_csv('../Dataset/PL-teams.csv')
```

```python
# Filtering Dataset keeping details before May 25th, 2025
df_history['match_date'] = pd.to_datetime(df_history['match_date'])
df_history = df_history[df_history['match_date'] <= '2025-05-25']
```

```python
# Dropping some unwamted columns
cols_to_drop = [
    'match_id', 'home_team_id', 'away_team_id', 'outcome_id', 'outcome_label',
    'forecast_home_win', 'forecast_draw', 'forecast_away_win', 
    'market_home_edge', 'market_expected_points_home', 'market_expected_points_away', 
    'market_entropy', 'market_logit_home', 'market_max_prob'
]
df_history = df_history.drop(columns=cols_to_drop, errors='ignore')
```

```python
# Checking everything on recent dataset
e0_essential_cols = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']
df_recent = df_recent[e0_essential_cols]
```

```python
# Renaming columns to fit our previous dataset
df_recent = df_recent.rename(columns={
    'Date': 'match_date',
    'HomeTeam': 'home_team_name',
    'AwayTeam': 'away_team_name',
    'FTHG': 'home_goals',
    'FTAG': 'away_goals',
    'FTR': 'match_outcome'
})
```

```python
# Adjusting Dates to match our dataset
df_recent['match_date'] = pd.to_datetime(df_recent['match_date'], dayfirst=True)
```

```python
# Combinig the trimed datas with our new data
df_master = pd.concat([df_history, df_recent], ignore_index=True)
```

```python
# Sorting the dataset
df_master = df_master.sort_values(by='match_date').reset_index(drop=True)
```

```python
# Keeping the same team name across all combined datasets
team_replacements = {
    'Man City': 'Manchester City',
    'Man United': 'Manchester United',
    'Nott\'m Forest': 'Nottingham Forest',
    'Spurs': 'Tottenham Hotspur'
}
df_master['home_team_name'] = df_master['home_team_name'].replace(team_replacements)
df_master['away_team_name'] = df_master['away_team_name'].replace(team_replacements)
```

```python
# Merging home-team details 
df_master = df_master.merge(
    df_ea_teams, 
    left_on='home_team_name', 
    right_on='Team Name', 
    how='left'
)
```

```python
# Rename new ea columns so we know where they belong
df_master = df_master.rename(columns={
    'Attack': 'home_ea_attack',
    'Midfield': 'home_ea_midfield',
    'Defense': 'home_ea_defense',
    'Overall': 'home_ea_overall'
}).drop(columns=['Team Name'])
```

```python
# Same for away team
df_master = df_master.merge(
    df_ea_teams, 
    left_on='away_team_name', 
    right_on='Team Name', 
    how='left'
)
df_master = df_master.rename(columns={
    'Attack': 'away_ea_attack',
    'Midfield': 'away_ea_midfield',
    'Defense': 'away_ea_defense',
    'Overall': 'away_ea_overall'
}).drop(columns=['Team Name'])
```

```python
# Exporting our final dataset
df_master.to_csv('../Dataset/Master_Capstone_Dataset.csv', index=False)
print(f"Total Matches: {len(df_master)} | Columns: {len(df_master.columns)}")
```

```python
df = pd.read_csv('../Dataset/Master_Capstone_Dataset.csv')
```

```python
# Drop some columns
cols_to_drop = ['match_datetime_utc', 'match_weekday'] + [c for c in df.columns if c.endswith('.1')]
df = df.drop(columns=cols_to_drop, errors='ignore')
```

```python
# Fixing basic math flags and recalculating them
df['total_goals'] = df['home_goals'] + df['away_goals']
df['goal_difference'] = df['home_goals'] - df['away_goals']

df['home_win_flag'] = (df['home_goals'] > df['away_goals']).astype(int)
df['away_win_flag'] = (df['home_goals'] < df['away_goals']).astype(int)
df['draw_flag'] = (df['home_goals'] == df['away_goals']).astype(int)

df['home_points_actual'] = df['home_win_flag'] * 3 + df['draw_flag'] * 1
df['away_points_actual'] = df['away_win_flag'] * 3 + df['draw_flag'] * 1
```

```python
# Standarizing outcome to H,A and D
def standardize_outcome(row):
    if row['home_goals'] > row['away_goals']: return 'Home Win', 'H'
    elif row['home_goals'] < row['away_goals']: return 'Away Win', 'A'
    else: return 'Draw', 'D'

outcomes = df.apply(standardize_outcome, axis=1)
df['match_outcome'] = [x[0] for x in outcomes]
df['match_outcome_code'] = [x[1] for x in outcomes]
```

```python
# Addressing XG gap on the merged dataset 
df['home_xg'] = df['home_xg'].fillna(df['home_goals'])
df['away_xg'] = df['away_xg'].fillna(df['away_goals'])
df['xg_difference'] = df['home_xg'] - df['away_xg']
```

```python
# Fixing EA ratings and relegated teams and filling missing columns
df_ea = pd.read_csv('../Dataset/PL-teams.csv')
ea_dict_overall = dict(zip(df_ea['Team Name'], df_ea['Overall']))
ea_dict_attack = dict(zip(df_ea['Team Name'], df_ea['Attack']))
ea_dict_mid = dict(zip(df_ea['Team Name'], df_ea['Midfield']))
ea_dict_def = dict(zip(df_ea['Team Name'], df_ea['Defense']))
```

```python
# As we have mismatched names in the combined dataset as we dont have team ratings for recently promoted teams so by default we are giving them 75
team_mapping = {
    'Brighton': 'Brighton & Hove Albion',
    'Ipswich': 'Ipswich Town',
    'Leicester': 'Leicester City',
    'Newcastle': 'Newcastle United',
    'Tottenham': 'Tottenham Hotspur',
    'West Ham': 'West Ham United',
    'Wolves': 'Wolverhampton Wanderers'
}

def get_rating(team, rating_dict):
    mapped_team = team_mapping.get(team, team)
    return rating_dict.get(mapped_team, 75.0) 

df['home_ea_overall'] = df['home_team_name'].apply(lambda x: get_rating(x, ea_dict_overall))
df['home_ea_attack'] = df['home_team_name'].apply(lambda x: get_rating(x, ea_dict_attack))
df['home_ea_midfield'] = df['home_team_name'].apply(lambda x: get_rating(x, ea_dict_mid))
df['home_ea_defense'] = df['home_team_name'].apply(lambda x: get_rating(x, ea_dict_def))

df['away_ea_overall'] = df['away_team_name'].apply(lambda x: get_rating(x, ea_dict_overall))
df['away_ea_attack'] = df['away_team_name'].apply(lambda x: get_rating(x, ea_dict_attack))
df['away_ea_midfield'] = df['away_team_name'].apply(lambda x: get_rating(x, ea_dict_mid))
df['away_ea_defense'] = df['away_team_name'].apply(lambda x: get_rating(x, ea_dict_def))
```

```python
# Dropping other columns from merged dataset
df_clean = df.dropna(axis=1)

df_clean.to_csv('../Dataset/Cleaned_Master_Dataset.csv', index=False)
print(f"Final ML-Ready Columns Left: {len(df_clean.columns)}")
print(df_clean.columns.tolist())
```
